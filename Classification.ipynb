## Working with categorical features

```python
# Convert each column dtype to object
for col in df.columns:
    if (col[:1]=='C') | (col[:1]=='T'):
        df[col] = df[col].astype(str)
    elif (col[:1]=='N') & (df[col].dtypes=='object'):
        df[col] = df[col].apply(lambda x: float(x.replace(',', '.')))
```

# Data preprocessing

## Data cleaning

### Missing values

```python
# Print the number of null values in each column
df.isnull().sum()
```

## Feature selection

### Columns with high cardinality

```python
# Because C1 is a unique variable it will not have so important impact on prediction
# Then we drop it from data
df.drop('C1', axis=1, inplace=True)
```

For those kind of variables we will use **Label encoding**. Because all values already labeled we just convert the column type to integer

```python
# C2
df['C2'] = df['C2'].astype(int)
dummy = pd.get_dummies(df[['C3']])
df['C3_31'] = dummy['C3_31']
df['C3_32'] = dummy['C3_32']
df.drop('C3', axis=1, inplace=True)
```

### Columns with large amount of zeros

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(df[['N1']])
df['N1'] = scaler.transform(df[['N1']])
```

```python
for col in df.columns:
    if (df[col].dtypes!='object') & (df[col].nunique()>10):
        scaler = StandardScaler().fit(df[[col]])
        df[col] = scaler.transform(df[[col]])
```
